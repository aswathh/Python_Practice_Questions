**Q1.Given an input string s, reverse the order of the words.**

```python

def reverseWords(s):
    words =s.split()
    words=words[::-1]
    return " ".join(words)

print(reverseWords("the sky is blue"))
```

**Q2.Given a string s consisting of words and spaces, return the length of the last word in the string.**

```python
def lengthOfLastWord(s):
        words = s.split()
        last = len(words[-1])
        return last
print(lengthOfLastWord("luffy is still joyboy"))
```

**Q3.Find the Index of the First Occurrence in a String**

```python
def strStr(haystack, needle):
    n = len(haystack)
    m = len(needle)
    for i in range((n-m)+1):
        if haystack[i:i+m] == needle:
            return i
    return -1

haystack ="hello"
needle ="ll"
print(strStr(haystack, needle))
```

**Q4.Palindrome sentence**

```python
def palindrome_sentence(s):
    phrase = "".join(c.lower() for c in s if c.isalnum())
    return phrase == phrase[::-1]
sd = "A man, a plan, a canal: Panama"
print(palindrome_sentence(sd))
```

**Q6.Two Sum**

```python
#Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
def two_sum(nums,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            return nums[i]+nums[j] == target
nums = [2,7,11,15]
target = 9

print(two_sum(nums,target))
```

**Q7. Merge Two Sorted Lists**

```python
def merge_list(l1,l2):
     l1.extend(l2)
     l1.sort()
     return l1
     
l1 = [1,2,4]
l2 = [1,3,4]
print(merge_list(l1,l2))
```

**Q8. Single Number**

```python
#Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.
def single_number(nums):
    ans=0
    for num in nums:
        ans ^= num
        return ans
nums=[4,1,2,1,2]
print(single_number(nums))
```

**Q9.missingNumber in list**

```python
# Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.
def mising_number_list(nums):
    n = len(nums)
    for i in range(n+1):
        if i not in nums:
            return i
nums = [9,6,4,2,3,5,7,0,1]
print(mising_number_list(nums))
```

**Q.10 Reverse Bits**

```python
#Reverse bits of a given 32 bits signed integer.
def reverse_bit(n):
    binary= bin(n)[2:]
    binary = binary.zfill(32)
    binary = binary[::-1]
    return int(binary,2)
n=43261596
print(reverse_bit(n))
```

**Q.12 Get the Size of a DataFrame**

```python
import pandas as pd
from typing import List
def create_dataframe(student_data:list[list[int]]) -> pd.DataFrame:
    return pd.DataFrame.from_records(student_data,columns=["student_id","age"])
def dataframe_size(student_data):
    return list(student_data.shape)
student_data=[
  [1, 15],
  [2, 11],
  [3, 11],
  [4, 20]
]
df = create_dataframe(student_data)
print(df)

print(dataframe_size(df))
```

**Q13.Longest Common Prefix**

```python
def common_prefix(lst)->str:
    prefix = lst[0]
    for s in lst[1:]:
        while not s.startswith(prefix):
            prefix= prefix[:-1]
            if not prefix:
                return ""
    return prefix       
lst =["flower","flow",,"flight"]
print(common_prefix(lst))

```
