# Interview Basic Questions

### Q1. Write a Python program to check if a number is even or odd.

```python
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
n = int(input("Enter a number: "))
print(even_or_odd(n))
```

### Q2. Write a Python function to reverse a string.

```python
def reverse_string(s)-> str:
    return s[::-1]
s = "Enter a number"
print(reverse_string(s))
```

### Q3. How can you find the largest number in a list?

```python
def largest_number(lst) -> int:
    return max(lst)
lst = [45,65,79,103,27,96]
print(largest_number(lst))
```

### Q4. Write a program to calculate the factorial of a number.

```python
def factorial(number):
    if number == 0:
        return 1
    else:
        return number*factorial(number-1)
    
factorial(9)
```

### Q5. How do you check if a string is a palindrome?

```python
def palindrome(string):
    if string == string[::-1]:
        return "palindrome"
    else:
        return "not_palindrome"
    
palindrome("goat")
```

```python
def palindrome(s):
    return s == s[::-1]
palindrome("malayalam")
```

### Q6. Write a program to count the number of vowels in a given string.

```python
def count_vowels(string):
    return sum(1 for char in string if char.lower() in 'aeiou')
count_vowels("alphabet")
```

### Q7. How do you remove duplicates from a list?

```python
def unique_list(lst):
    return list(set(lst))
unique_list([1,3,1,5,1,3,9,7])
```

### Q8. Write a program to find the common elements in two lists.1

```python
def common_elements(lst1,lst2):
    a = list(set(lst1)) and list(set(lst2))
    return a
common_elements([2,4,8,5,0,1,2,4],[3,4,5,6,9,2,3,4])
```
