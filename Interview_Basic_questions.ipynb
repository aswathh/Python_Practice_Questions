# Interview Basic Questions

### Q1. Write a Python program to check if a number is even or odd.

```python
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
n = int(input("Enter a number: "))
print(even_or_odd(n))
```

### Q2. Write a Python function to reverse a string.

```python
def reverse_string(s)-> str:
    return s[::-1]
s = "Enter a number"
print(reverse_string(s))
```

### Q3. How can you find the largest number in a list?

```python
def largest_number(lst) -> int:
    return max(lst)
lst = [45,65,79,103,27,96]
print(largest_number(lst))
```

### Q4. Write a program to calculate the factorial of a number.

```python
def factorial_num(number):
    if number == 0:
        return 1
    else:
        return number*factorial_num(number-1)
    
print(factorial_num(9))
```

### Q5. How do you check if a string is a palindrome?

```python
def palindrome(s)->bool:
    if s == s[::-1]:
        return "palindrome"
    else:
        return "not_palindrome"
s="myself"
print(palindrome(s))
```

```python
def palindrome(s):
    return s == s[::-1]
palindrome("malayalam")
```

### Q6. Write a program to count the number of vowels in a given string.

```python
def vowels_count(s)->int:
    vowels = "aeiou"
    return sum(1 for char in s if char in vowels)
vowels_count("amazon")
```

### Q7. How do you remove duplicates from a list?

```python
def unique(lst)-> list:
    return list(set(lst))
num = [30,40,50,33,43,64,30,40,50]
unique = sorted(unique(num))
print(unique)
```

### Q8. Write a program to find the common elements in two lists.1

```python
def common(lst1,lst2)->list:
    lst = set(lst1) & set(lst2)
    return sorted(list(lst))
lst1 = [10,50,40,20,17]
lst2 = [70,80,20,40,10]
print(common(lst1,lst2))
```

**Q9. How do you merge two dictionaries in Python?**

```python
def merge_dict(dict1,dict2)->dict:
    merged = {**dict1,**dict2}
    return merged
dict1 = {"a":5,"b":6}
dict2 ={"b":7,"c":8}
print(merge_dict(dict1,dict2))
```
