### Q1. Write a Python program to check if a number is even or odd.

```python
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
n = int(input("Enter a number: "))
print(even_or_odd(n))
```

### Q2. Write a Python function to reverse a string.

### Q3. How can you find the largest number in a list?

### Q4. Write a program to calculate the factorial of a number.

```python
def factorial_num(number):
    if number == 0:
        return 1
    else:
        return number*factorial_num(number-1)
    
print(factorial_num(9))
```

### Q5. How do you check if a string is a palindrome?

```python
def palindrome(s)->bool:
    if s == s[::-1]:
        return "palindrome"
    else:
        return "not_palindrome"
s="myself"
print(palindrome(s))
```

### Q6. Write a program to count the number of vowels in a given string.

```python
def vowels_count(s)->int:
    vowels = "aeiou"
    return sum(1 for char in s if char in vowels)
vowels_count("amazon")
```

### Q7. How do you remove duplicates from a list?

```python
def unique(lst)-> list:
    return list(set(lst))
num = [30,40,50,33,43,64,30,40,50]
unique = sorted(unique(num))
print(unique)
```

**Q9. How do you merge two dictionaries in Python?**

```python
def merge_dict(dict1,dict2)->dict:
    merged = {**dict1,**dict2}
    return merged
dict1 = {"a":5,"b":6}
dict2 ={"b":7,"c":8}
print(merge_dict(dict1,dict2))
```

**Q11. How can you count the occurrences of each character in a string?**


```python
from collections import Counter
def occurrences_char(s)-> int:
    return dict(Counter(s))
print(occurrences_char("malayalam"))
```

**Q12. Write a program to find the Fibonacci series up to n.**

```python
def fibonacci_series(n)-> list:
    fibonacci =[0,1]
    while len(fibonacci) < n:
        fibonacci.append(fibonacci[-1]+fibonacci[-2])
    return fibonacci[:n]
n= int(input("Enter a number: "))
print(fibonacci_series(n))
```

**Q14. How do you convert a list of strings to uppercase?**

**Q15. How do you find the length of a list without using the len() function?**

```python
def len(lst)-> int:
    count =0
    for i in lst:
        count +=1
    return count
lst =[10,60,40,50,45,63,72]
print(len(lst))
```

**Q16. Write a Python program to check whether two strings are rotations of each other.**

```python
def rotation_string(s1,s2)-> bool:
    return len(s1)==len(s2) and s2 in(s1+s1)
s1 ="water"
s2 = "terwa"
print(rotation_string(s1,s2))
```

**Q17. Write a Python function to find duplicate elements in a list.**

```python
def find_duplicates(lst)-> list:
    seen=set()
    duplicates=set()
    for item in lst:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)
n = [20,34,40,30,10,20,50,60,40,20,10]
print(sorted(find_duplicates(n)))
    
```

**Q18. Write a Python function to check if a number is prime.**

```python
def is_prime(num)->str:
    if num <=1:
        return "Not a Prime Number !!"
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            return "Not a Prime Number!!"
    return "Prime Number!!"
print(is_prime(457))
```

**Q19. Write a program to find the second largest number in a list.**

```python
def second_largest(lst)-> int:
    lst1 = list(set(lst))
    sorted_lst = sorted(lst1)
    return sorted_lst[-2]
lst =[2,9,10,30,56,100,100,110,110,120,60,30,120,48,45,36,75,99]
print(second_largest(lst))
```
