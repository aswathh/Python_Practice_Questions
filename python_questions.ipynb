```python
# Write a function is_prime(n) that returns True/False if a number is prime.
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
    
print(is_prime(2))
print(is_prime(13))
print(is_prime(15))
```

## OOP (Classes & Inheritance) + NumPy/Pandas Coding Round

```python
# Class = a blueprint. Object = the actual thing built from that blueprint
class Employee: # class
    def __init__(self,name,role): # constructor - runs when Employee object is created
        self.name =name # attribute
        self.role =role # self means => "Store this information inside ME."
    def introduce(self): #method
        return f"I am {self.name}, working as {self.role}"
emp1 = Employee("Aswath", "AI Engineer") #object
emp2 = Employee("Rahul", "Data Analyst")
print(emp2.introduce())

# Without self, Python wouldn't know whether to use Aswath's data or Rahul's data.
# self.role => stores it inside the object permanently.
# self = emp1 or emp2 => by which object we are creating that object assigned at self
```

```python
# Inheritance — child class reuses parent class:
class Employee:
    def __init__(self,name):
        self.name =name
    def work(self):
        return f"{self.name} is working"
class Manager(Employee): # Manager inherits Employee
    def __init__(self,name,team_size):
        super().__init__(name) # call parent constructor
        self.team_size =team_size
    def work(self):# method overriding
        return f"{self.name} is working with {self.team_size} members"
m1 = Manager("priya",8)
m2 = Manager("siva",9)
print(m2.work())
print(m1.work())
```

##### polymorphism

```python
# Create a base class Model with a method train(). Create two child classes LinearModel and TreeModel that override train() differently.
class Model:
    def train(self):
        print("Generic training process")
class LinearModel(Model): # inherit
    def train(self):# method overriding
        print("Training using gradient descent")
class TreeModel(Model): # inherit
    def train(self):# method overriding
        print("Training using splittinng nodes")

Models =[LinearModel(),TreeModel()]
for m in Models:
    m.train() #object

#This pattern (polymorphism) is exactly how scikit-learn's .fit() works under the hood across different model classes!
```

```python
# Create a class BankAccount with deposit() and withdraw() methods, and an attribute balance. Prevent withdrawal if balance is insufficient.
class BankAccount:
    def __init__(self,balance=0):
        self.balance =balance
    
    def deposit(self,amount):
        self.balance +=amount
        print(f"Dposited {amount}, New balance:{self.balance}")
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance-=amount
            print(f"Withdraw {amount}, New balance:{self.balance}")
acc= BankAccount(1000)
acc.deposit(500)
acc.withdraw(3000)
    
```

## File Handling, Exceptions, Modules + LLM/GenAI/RAG Concepts


##### File handling — like reading/writing a log file during deployment:

```python
# write a file 
with open("log.txt","w") as f:
    f.write("Deployment started\n")
    f.write("Status: Success\n")
    f.write("ERROR:ZeroDivisionError\n")

# read a file 
with open("log.txt","r") as f:
    content = f.read()
print(content)    

# read line by line
with open("log.txt","r") as f:
    for line in f:
        print(line.strip())

# with automatically closes the file — even if an error happens. Always prefer this over manual open()/close().
```

##### Exception Handling — real data/APIs are messy, this is non-negotiable in production code:

```python
try:
    age = int(input("enter a number:"))
    result = 100/age
except ValueError:
    print("Enter a valid number")
except ZeroDivisionError:
    print("age must not be zero")
else:
    print("Result:",result)
finally:
    print("Execution succeeded")

```

```python
# Scenario: You're calling an external API (e.g., OpenAI/Bedrock) in a loop for 100 requests. One request fails due to a timeout. How do you prevent the whole loop from crashing?
import time

result =[]
for i in range(100):
    try:
        # response = call_llm_api(prompt[i])  # imagine API call
        response= "suceess"
        result.append(response)
    except TimeoutError:
        print(f"Request {i} timed out, retrying...")
        time.sleep(2)
    except Exception as e:
        print(f"Request {i} failed: {e}")
        continue
# This exact pattern shows up in real GenAI pipelines — handling rate limits, timeouts, and malformed responses.
```
